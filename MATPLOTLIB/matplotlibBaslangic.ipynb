```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#şimdi yas ile kilo arasında bir liste oluşturalım. ondan sonra matplotlib grafik oluşturalım.
yasListesi= [10,12,19,21,54,87,92,23,21,45]
```

```python
kiloListesi=[20,36,25,47,85,14,36,98,46,80]
```

```python
#şimdi bu listeleri numpy listelerine çevirelim.
numpyYasListesi= np.array(yasListesi)
numpyKiloListesi= np.array(kiloListesi)

#şimdi de plot işlemi, biz bunun hangi şekilde ve hangi renkte olduğunu da belirtiyoruz.
plt.plot(numpyYasListesi,numpyKiloListesi,"g") #burdaki "g" aslında yeşil
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r") 


#peki ben bunlara eksen adı verebilir miydim ? Bunu plt.xlabel fonksiyonu ile yapıcam.


plt.xlabel("Yas")
plt.ylabel("Kilo")

plt.title("Kilonun Yasa göre değişim grafiği")
#plt.show 
```

```python
#biraz da numpy kütüphanesinin nimetlerinden yararlanalım. 

numpyDizisiNormal= np.linspace(0,10,20) #bu linspace fonksiyonu bizim için belli aralıkla bi dizi oluşturuyordu.
#yukarı 0 ile 10 arasında 20 aralıklı bir dizi oluşturdum
```

```python
numpyDizisiNormal
```

```python
numpyDizisiUstel= numpyDizisiNormal **3
```

```python
plt.plot(numpyDizisiNormal,numpyDizisiUstel,"b*-")
```

```python
#aynı düzlemde iki farklı grafik oluşturma

plt.subplot(1,2,1) #yanyana da alt alta grafik gösterebilrim. İlk parametrem kaç sıra olacağı,ikinci parametrem kaç grafik göstereceğim?,şimdi hangisini çizeceksem 3. parametrem o oluyor.
plt.plot(numpyDizisiNormal,numpyDizisiUstel,"r*-")

plt.subplot(1,2,2) #1 sıra olacak, 2 kolon olacak,2. grafiği çiziyorum

plt.plot(numpyDizisiUstel,numpyDizisiNormal,"b--")
```

```python
#özelleştirilmiş grafikler yapamaz mıyız ? bunun için plt.figure fonksiyonundan yardım alıcaz.

```

```python
#yeni bir figure oluşturmak için

benimFigure=plt.figure()
figureAxes= benimFigure.add_axes([0.4,0.4,0.6,0.6]) #width ve height bunlar. x1,x2,y1,y2 şeklinde. x1-x2/y1-y2
figureAxes.plot(numpyDizisiNormal,numpyDizisiUstel,"r-*")

figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")


figureAxes.set_title("Benim Grafiğim")

```

```python
#şimdi başka bir figure yapıcaz mesafeleri daha iyi anlamak için

figure2= plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.1,0.3,0.3,0.3])

eksen1.plot(numpyDizisiNormal,numpyDizisiUstel,"b")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")



eksen2.plot(numpyDizisiUstel,numpyDizisiNormal,"r")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
#şimdi plt.figure ile plt.subplots arasındaki farklardan biri de subplot'da eksenlerin birlikte gelmesi ancak
#sadece 

plt.figure() 
#şu inputu almış oluyorum <Figure size 432x288 with 0 Axes>

#bundan dolayıdır ki plt.subplots içinde bana farklı değişkenler döndürmüş oluyor.

(benimFİGURUM,benimEksenler) = plt.subplots(nrows=1,ncols=2) #şeklinde de düşünebiliriz.,numberrows--numbercoloumns :)

type(benimEksenler)
```

```python
(benimFİGURUM,benimEksenler) = plt.subplots(nrows=1,ncols=2) #şeklinde de düşünebiliriz.,numberrows--numbercoloumns :)

type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisiNormal,numpyDizisiUstel,"g")
    eksen.set_xlabel("X EKSEN")


plt.tight_layout()

```

```python
#bu sefer bu grafiklerde ne tarz değişikler yapabiliriz. onu öğrenicez.

yeniFigur=plt.figure(dpi=100) #dpi bizim için önemli bir parametre dpi (dot per int demek yani pixels anlamına geliyor) #figsize aslında biz axeslerden hallediyoruz.. 

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisiNormal,numpyDizisiUstel,label ="numpyDizisi **2 ")
yeniEksen.plot(numpyDizisiNormal,numpyDizisiNormal**9,label ="numpyDizisi **9 ")
#yeniEksen.legend() #legend== ne verisi olduğunu anlamına geliyor.+

#peki ben bu legend fonksiyonunu başka yere koyabilir miydim ? 

yeniEksen.legend(loc=2)
```
